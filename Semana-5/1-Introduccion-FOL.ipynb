# Logica de Primer Orden

La lógica de primer orden, como todos los demás sistemas de lógica formal, es un método para formalizar lenguajes naturales en un formato computable. Esto, a su vez, nos permite tratar los problemas expresados a través de oraciones lingüísticas de una manera formal. De la formalización del lenguaje natural, derivamos la capacidad de formular y probar teoremas, lo que a su vez permite llevar a cabo el razonamiento inferencial en disciplinas como las matemáticas, la informática y la filosofía.

El problema de la formalización de los lenguajes naturales es tan antiguo como la filosofía misma. En el campo de la informática, sin embargo, encuentra sus bases en la década de 1950 cuando se desarrollaron los primeros sistemas para el procesamiento de expresiones lógicas. Sin embargo, el primero de estos sistemas tenía severas limitaciones y, en particular, solo podían trabajar con lógica proposicional pero no predicada.

La lógica proposicional no permite la realización del razonamiento sobre variables y funciones de contenido general y mutable, lo que significa que su capacidad de abstracción es limitada. Esto significa también que los primeros sistemas lógicos de computación no podían resolver problemas cuya solución radica en espacios vectoriales de los cuales el espacio proposicional es un subespacio. Este problema se superó gracias al desarrollo de un sistema lógico formal, la lógica de primer orden, que incluye variables y, por lo tanto, permite la abstracción.

## Logica Proposicional vs Logica de Primer Orden

- La primera diferencia entre los dos se relaciona con el hecho de que la lógica de primer orden incluye la lógica proposicional, pero lo contrario no es cierto. Esto significa que todos los problemas expresados en lógica proposicional pueden ser tratados bajo lógica de primer orden, pero no todos los problemas en lógica de primer orden pueden ser tratados en lógica proposicional. Esto se refiere, en particular, a problemas que solo pueden expresarse haciendo uso de cuantificadores, como "todos", "cualquiera", "algunos", "ninguno", como veremos más adelante con más detalle.
- La segunda diferencia se refiere a la naturaleza de la unidad elemental que constituye las fórmulas de los dos sistemas formales. La lógica proposicional utiliza proposiciones y operadores lógicos para constituir sus fórmulas bien formadas. La lógica de primer orden, además de esos, también utiliza variables, cuantificadores y relaciones. Vamos a ver en las siguientes secciones lo que significan estos nuevos conceptos.
- La tercera diferencia se refiere a la capacidad de abstracción de fórmulas expresadas por los dos sistemas. En lógica proposicional, el sistema no permite manejar problemas que involucran partes mutantes o indeterminadas. Esto significa que la validez de una solución a una fórmula que se encuentra en la lógica proposicional se restringe a esa fórmula. En la lógica de primer orden, por el contrario, es posible crear fórmulas que posean una mayor capacidad de generalización.

## Componentes de la Lógica de Primer Orden
 
-	Objetos: P, Q, Juan, Anna, Wumpus, etc
-	Relaciones: existen relaciones unarias y n-arias como Juan-[profe]-David, donde “profe” es el nombre de la relación.
-	Funciones: pueden devolver true o false o una lista de elementos. Por ejemplo, EsProfe(Juan) devuelve True.
-	Sintaxis: es muy similar a la de la lógica proposicional pero existen elementos nuevos como predicados, y quantificadores como (∀, ∃)
-	Semántica: al igual que en la lógica proposicional existen formas de declarar oraciones atómicas y complejas. Vamos a ver algunos ejemplos a continuación.

**Oraciones Atomicas**: Determinan un hecho!

- Brother(Richard ,John). "Richard corazon de leon es el hermano del rey John"
- Married(Father (Richard),Mother (John)). "El padre de Richard corazon de leon esta casado con la madre del rey John".

**Oraciones Complejas**

- ¬Brother (LeftLeg(Richard), John)
- Brother (Richard , John) ∧ Brother (John,Richard)
- King(Richard) ∨ King(John)
- ¬King(Richard) ⇒ King(John).

**Quantificadores**

- Universal: ∀x King(x) ⇒ Person(x) "Para todo x, si x es rey, entonces x es una persona"
- Existencial: ∃x Crown(x) ∧ OnHead(x, John) "Existe al menos un x, donde x es una corona y la corona esta en la cabeza de
John"

**Quantificadores Anidados**

- ∀x∀y Brother(x, y) ⇒ Sibling(x, y) "Los hermanos son parientes"
- ∀x,y Sibling(x, y) ⇔ Sibling(y, x) "los parientes tienen una relacion simetrica"
- ∀x∃y Loves(x, y) "Todo el mundo ama a alguien (al menos a una persona!)"
- ∃y∀x Loves(x, y) "Existe alguien a quien lo aman todos"
- ∀x (Crown(x) ∨ (∃x Brother(Richard, x)))

**Conexiones entre ∀ e ∃**

- ∀x ¬Likes(x, Parsnips ) es equivalente a ¬∃x Likes(x, Parsnips)
- ∀x Likes(x, IceCream) es equivalente a ¬∃x ¬Likes(x, IceCream)
