# Demostración de Teoremas de Lógica Proposicional

Existen diversas formas de verificar la implicación lógica. Ya vimos como pudimos hacerlo usando el chequeo de modelos, es decir, enumerar los modelos de KB y ver si la sentencia se sostiene a través de todos los modelos.

Hay otra forma de realizar esto, por medio de “theorem proving”, es decir, aplicando reglas de inferencia directamente a las oraciones en nuestra base de conocimiento para construir una prueba de la oración deseada sin consultar modelos.

## Equivalencia Lógica

α≡ β si y solo si α |= β y β |= α .


## Validez: teorema deductivo

Para cualquier oración α y β, α |= β si y solo si la oración (α ⇒ β) es valida.

Ósea, podemos decidir si α |= β  al probar que (α ⇒ β)  en cada modelo.

Por ejemplo dada la oración: 

α |= β si y solo si la oración (α ∧ ￢β) es insatisfactorio.

Probar β a partir de α comprobando la insatisfactoria de (α ∧ ¬β) corresponde exactamente a la técnica de demostración matemática estándar de “reductio ad absurdum” (literalmente, "reducción a una cosa absurda"). También se llama prueba por refutación o prueba por contradicción.

## Inferencia y Pruebas

**Modus Ponens** 

$$\frac{α ⇒ β, α}{β}$$

La notación significa que, siempre que se dan oraciones de la forma  α ⇒ β y α, entonces se puede inferir la oración β. Por ejemplo, si (WumpusAhead ∧ WumpusAlive) ⇒ Disparar y (WumpusAhead ∧ WumpusAlive) se dan, luego se puede inferir Shoot.

**And-Elimination** 

$$\frac{α ∧ β}{α}$$

Por ejemplo, de (WumpusAhead ∧ WumpusAlive) se puede inferir WumpusAlive.

## Aplicación en Wumpus World

Vamos a utilizar las reglas de R1 a R5 para probar ￢P1,2.

    Se aplica eliminación bicondicional a R2 para obtener R6:
    R6:  (B1,1 ⇒ (P1,2 ∨ P2,1)) ∧ ((P1,2 ∨ P2,1) ⇒ B1,1)

    And-Elimination a R6 para obtener R7:
    R7: ((P1,2 ∨ P2,1) ⇒ B1,1)

    Equivalencia logica para contrapositivas:
    R8: (￢B1,1 ⇒ ￢(P1,2 ∨ P2,1))

    Modus Ponens a R8
    R9 : ￢(P1,2 ∨ P2,1)

    Ley DeMorgan:
    R10 : ￢P1,2 ∧ ￢P2,1

    R10 indica que ni [1,2] o [2,1] son Pozos (pits) 

Realizar estas pruebas a mano requiere conocimiento matemático, pero se puede automatizar utilizando los algoritmos de búsqueda que hemos visto. 

Por lo tanto, la búsqueda de pruebas es una alternativa a la enumeración de modelos. En muchos casos prácticos, encontrar una prueba puede ser más eficiente porque la prueba puede ignorar proposiciones irrelevantes, sin importar cuántas de ellas haya.
