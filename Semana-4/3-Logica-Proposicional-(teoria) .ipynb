# Lógica Proposicional 

Este es el primer tipo de lógica que vamos a usar para nuestro agente “lógico”.  La sintaxis de la lógica proposicional define las oraciones permitidas. Las oraciones atómicas consisten en un solo símbolo de proposición. Cada uno de estos símbolos representa una proposición que puede ser verdadera o falsa. Algunos ejemplos de símbolos son P, Q , R o bien W13 y Norte. Existen 2 símbolos que siempre significan lo mismo: True y False.

## Los 5 conectivos más comunes:

-	￢ (No o negación): este permite negar una oración. Por ejemplo A = El wumpus está muerto, entonces ￢A significa que el wumpus está vivo.
-	∧ (and o conjunción): Suele indicar que dos oraciones son verdaderas, por ejemplo (P ∧ A)
-	∨ (or o disyunción): indica que solo una oración es verdadera; por ejemplo (P ∨ Q) solo una es cierta.
-	⇒ (implica, if-then): esto es usado para indicar que una oración es un antecedente de otro. Por ejemplo (P ∧ Q)  ⇒ A. Esto quiere decir que si P y Q son verdaderas, entonces ocurre A. Puede ver esto como un if-then; if (P and Q) then A. 
-	 ⇔ (si y solo si, bicondicional, iff): La oración P ⇔ A indica que P es verdadera solo si A es verdadera.

## Semántica

La semántica define las reglas para determinar la verdad de una oración con respecto a un modelo particular. En la lógica proposicional, un modelo simplemente fija el valor de verdad, verdadero o falso, para cada símbolo de proposición.

Modelo m1 = {P12 = false, P22 = false, P31 = true}

La semántica para la lógica proposicional debe especificar cómo calcular el valor de verdad de cualquier oración, dado un modelo. Esto se hace de forma recursiva. Todas las oraciones se construyen a partir de oraciones atómicas y los cinco conectivos; por lo tanto, necesitamos especificar cómo calcular la verdad de las oraciones atómicas y cómo calcular la verdad de las oraciones formadas con cada una de las cinco conectivas.
