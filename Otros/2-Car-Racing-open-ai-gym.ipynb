La tarea de control continuo más fácil de aprender de los píxeles, un entorno de carreras de arriba hacia abajo. El control discreto también es razonable en este entorno, la discretización de encendido/apagado está bien. El estado consta de 96x96 píxeles. La recompensa es de -0,1 cada cuadro y +1000/N por cada mosaico de pista visitado, donde N es el número total de mosaicos en la pista. Por ejemplo, si ha terminado en 732 fotogramas, su recompensa es 1000 - 0,1*732 = 926,8 puntos. El episodio termina cuando se visitan todas las fichas. Algunos indicadores que se muestran en la parte inferior de la ventana y el búfer RGB de estado. De izquierda a derecha: velocidad real, cuatro sensores ABS, posición del volante, giroscopio.

Acciones:

- Girar Izquierda:  [ -1.0, 0.0, 0.0 ]
- Girar Derecha: [ +1.0, 0.0, 0.0 ]
- Frenar: [ 0.0, 0.0, 0.8 ]
- Acelerar: [ 0.0, 1.0, 0.8 ]
- Nada: [ 0.0, 0.0, 0.0 ]

```python
import pyvirtualdisplay
import gym
import matplotlib.pyplot as plt
from IPython import display
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

# use False con Xvfb 
_display = pyvirtualdisplay.Display(visible=False, size=(1400, 900))
_ = _display.start()

# Se crea el ambiente
env = gym.make("CarRacing-v0")
```

```python
import random 

observation = env.reset()
steps = 500
frames = []

print(observation.shape)

# Desplegar en pantalla (captura inicial del ambiente)
#img = plt.imshow(env.render(mode='rgb_array'))

for t in range(steps):

    # tome una accion aleatoria...
    action = [ 0.0, 1.0, 0.0 ]

    observation, reward, done, info = env.step(action)

    frame = env.render(mode='rgb_array')
    frames.append(frame)

    #print("Observaciones:", observation)
    #print("Reward:", reward)

    # Pantalla
    #img.set_data(env.render(mode='rgb_array')) 
    #plt.axis('off')
    #display.display(plt.gcf())
    #display.clear_output(wait=True)

    if done:
        print("Termino!")
        break
```

```python
# se crea una animacion y se despliega como un video.

from matplotlib.animation import FuncAnimation

def update_scene(num, frames, patch):
    patch.set_data(frames[num])
    return patch,

def plot_animation(frames, repeat=False, interval=40):
    fig = plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')
    anim = FuncAnimation(
        fig, update_scene, fargs=(frames, patch),
        frames=len(frames), repeat=repeat, interval=interval)
    plt.close()
    return anim

anim = plot_animation(frames)

from IPython.display import HTML
HTML(anim.to_html5_video())
```
