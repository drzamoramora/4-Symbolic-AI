# ¿Que es Inteligencia Artificial?
Segun patrones que buscan semejar a los humanos vs ejecutar tareas mejor que los humanos (orientadas a rendimiento).

### Pensar como Humano

* El nuevo y emocionante esfuerzo para hacer que las computadoras piensen... máquinas con mente, en el sentido pleno y literal. (Haugeland, 1985)
* La automatización de actividades que asociar con el pensamiento humano, las actividades tales como la toma de decisiones, resolución de problemas, aprendizaje… (Bellman, 1978)

### Actuar como Humano

* El arte de crear máquinas que realizan funciones que requieren inteligencia cuando son realizadas por personas. (Kurzweil, 1990)

### Pensar Racionalmente

* El estudio de las facultades mentales a través del uso de modelos computacionales. (Charniak y McDermott, 1985)
* El estudio de los cálculos que hacen posible percibir, razonar y actuar. (Winston, 1992)

### Actuar Racionalmente

* La Inteligencia Computacional es el estudio del diseño de agentes inteligentes. (Poole et al., 1998)
* IA... se ocupa del comportamiento inteligente en los artefactos. (Nilsson, 1998)

# La Prueba de Turing

<img src="img/alan.png" />

La prueba de Turing, propuesta por Alan Turing (1950), fue diseñada para proporcionar una definición operativa satisfactoria de inteligencia. Una computadora pasa la prueba si un interrogador humano, después de plantear algunas preguntas escritas, no puede saber si las respuestas escritas provienen de una persona o de una computadora.

Esta prueba sugiere la existencia de algunas capacidades como: 

* **Procesamiento del lenguaje natural** para permitirle comunicarse con éxito en inglés.
* **Representación del conocimiento** para almacenar lo que sabe o escucha
* **Razonamiento automatizado** para usar la información almacenada para responder preguntas y sacar nuevas conclusiones
* **Aprendizaje automático** para adaptarse a nuevas circunstancias y detectar y extrapolar patrones
* **Visión artificial** para percibir objetos
* **Robótica** para manipular objetos y moverse

# Ciencia Cognitiva

<img src="img/body.png" />

El campo interdisciplinario de la ciencia cognitiva reúne modelos informáticos de IA y técnicas experimentales de la psicología para construir teorías precisas y comprobables de la mente humana. 

En los primeros días de la IA, a menudo había confusión entre los enfoques: un autor argumentaría que un algoritmo se desempeña bien en una tarea y que, por lo tanto, es un buen modelo del desempeño humano, o viceversa.

# Las Bases de la Inteligencia Artificial

### Filosofía
* ¿Se pueden usar reglas formales para sacar conclusiones válidas?
* ¿Cómo surge la mente de un cerebro físico?
* ¿De dónde viene el conocimiento?
* ¿Cómo conduce el conocimiento a la acción?

### Matemáticas
* ¿Cuáles son las reglas formales para sacar conclusiones válidas?
* ¿Qué se puede calcular?
* ¿Cómo razonamos con información incierta?

### Ciencias económicas
* ¿Cómo debemos tomar decisiones para maximizar la rentabilidad?
* ¿Cómo debemos hacer esto cuando es posible que otros no estén de acuerdo?
* ¿Cómo deberíamos hacer esto cuando la recompensa puede estar muy lejos en el futuro?

### Neurociencia
* ¿Cómo procesa la información el cerebro?

### Psicología
* ¿Cómo piensan y actúan los humanos y los animales?

### Ingeniería Informática
* ¿Cómo podemos construir una computadora eficiente?

### Teoría del Control y Cibernética
* ¿Cómo pueden operar los artefactos bajo su propio control?

### Lingüística
* ¿Cómo se relaciona el lenguaje con el pensamiento?

# Fechas Importantes

* 1943-1955: Donald Hebb desarrolla el concepto del refuerzo de una conexion entre 2 neuronas. A esto se le llama "Hebian Learning"
* 1950: Prueba de Turing
* 1956: McCarthy y Misnky organizan el workshop de AI en Dartmouth en el verano del 56 con 10 personas entre estos Arthur Samuel.
* 1952-1969: Newell y Simon proponen el uso de sistemas de simbolos para que los sistemas inteligentes puedan tomar accion. En el 58 McCarthy desarrolla Lisp. En el 59, Herbert Gelernter desarrolla un probador de teoremas geometricos.
* 1966-1973: Primeras dificultades en aplicacion y uso de AI. Se cancela fondos para la traduccion automatica del Ruso al Ingles por parte del gobierno de EEUU. Otra dificultad que existio (y existe) es la tractabilidad de los problemas de AI: los sistemas servian en escenarios pequenos, pero fallaban en escenarios reales con mas complejidad. Otro problema reside en las dificultades de usar AI con las estructuras de datos del momento. Se desarrolla el Pereceptron (1969) como consequencia para aprender "cualquier cosa" pero con poco exito.
* 1969-1979: Se desarrolla el concepto de sistemas expertos. 
* 1986: El regreso de las redes neuronales. Se desarrolla el algoritmo de Backpropagation. Los modelos neuronales se clasifican como corrientes "conexionistas" en contrastes a los modelos basados en simbolos (Symbolic AI!)
* 1987: IA adopta el metodo cientifico.
* 1995: Se crea el concepto de agentes inteligentes por Newell, Laird y Rosenbloom.
* 1997: Kasparov pierde ante la Deep Blue de IBM
* 2001: Datasets enormes y el problema de datos masivos y algoritmos.
* 2005: DARPA concurso de automoviles autonomos.
* 2006-2009: Geoff Hinton et al, publican que ua red neuronal profunda puede ser pre-entrenada una capa a la vez y luego depurada con Back propagation.

# La Singularidad

La singularidad de la IA se refiere a un evento en el que las IA en nuestras vidas se vuelven conscientes de sí mismas o alcanzan una capacidad de mejora continua tan poderosa que evolucionará más allá de nuestro control.

<img src="https://static.probusiness.io/n/02/0/portrait_hi_tech_edit.jpg" />

El futurista Ray Kurzweil predijo hace 15 años que la singularidad, el momento en que las habilidades de una computadora superarán las habilidades del cerebro humano, ocurrirá alrededor de 2045. https://en.wikipedia.org/wiki/Ray_Kurzweil

# Tipos de Inteligencia Artificial

**Máquinas Reactivas**

Los tipos más básicos de sistemas de IA son puramente reactivos y no tienen la capacidad de formar recuerdos ni utilizar experiencias pasadas para informar las decisiones actuales. Deep Blue, la supercomputadora de ajedrez de IBM, que venció al gran maestro internacional Garry Kasparov a fines de la década de 1990, es el ejemplo perfecto de este tipo de máquina.


**Memoria Limitada**

Esta clase de Tipo II contiene máquinas que pueden mirar hacia el pasado. Los autos autonomos ya hacen algo de esto. Por ejemplo, observan la velocidad y la dirección de otros autos. Eso no se puede hacer en un solo momento, sino que requiere identificar objetos específicos y monitorearlos a lo largo del tiempo.

**Teoria de la Mente**

Podríamos detenernos aquí y llamar a este punto la importante división entre las máquinas que tenemos y las máquinas que construiremos en el futuro. 

Las máquinas de esta clase, más avanzada, no solo forman representaciones sobre el mundo, sino también sobre otros agentes o entidades en el mundo. En psicología, esto se llama "teoría de la mente": la comprensión de que las personas, las criaturas y los objetos del mundo pueden tener pensamientos y emociones que afectan su propio comportamiento.

**Auto-Conciencia**

El paso final del desarrollo de IA es construir sistemas que puedan formar representaciones sobre sí mismos. En última instancia, los investigadores de IA no solo tendremos que comprender la conciencia, sino también construir máquinas que la tengan.

"La singularidad ocurre aqui en esta etapa final, cuando se sobrepase la Inteligencia Artificial General, algo que no ocurrira en los proximos 50 años"
    - Dr. Zamora
